```python
import pandas_datareader.data as web
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# A1
ibm = web.get_data_yahoo("IBM", start='2016-01-01', end = '2020-12-31')['Adj Close'].pct_change().dropna()
print("최소 수익률 : ", ibm.min() * 100)
print("최대 수익률 : ", ibm.max() * 100)
print("평균 수익률 : ", ibm.mean() * 100)
```

```python
import pandas as pd
ibm_week = []
for i in range(len(ibm)):
    ibm_week.append(ibm[i * 5: (i * 5) + 5].sum())

ibm_week = pd.DataFrame(ibm_week)
ibm_week = ibm_week[ibm_week[0] != 0.000000]
```

```python
# A2
print("주간 최소 수익률 : ", ibm_week[0].min() * 100)
print("주간 최대 수익률 : ", ibm_week[0].max() * 100)
print("주간 평균 수익률 : ", ibm_week[0].mean() * 100)
```

```python
#A3
print("Annualized mean: " '%.3f%%' % (ibm_week[0].mean() * 252 * 100))
```

```python
sp500 = web.get_data_yahoo("^GSPC", start='2016-01-01', end='2020-12-31')['Adj Close'].pct_change().dropna()
sp500
```

```python
sp500_week = []
for i in range(len(sp500)):
    sp500_week.append(sp500[i * 5: (i * 5) + 5].sum())

sp500_week = pd.DataFrame(sp500_week)
sp500_week = sp500_week[sp500_week[0] != 0.000000]
sp500_week
```

```python
ibm_week = pd.Series(ibm_week[0])
sp500_week = pd.Series(sp500_week[0])
```

```python
import statsmodels.api as sm

Y = ibm_week
X = sm.add_constant(sp500_week)
model = sm.OLS(Y, X)
results = model.fit()
results.params
```

```python
results.summary()
```

B1 : 위의 코드에서 1.0614가 시장을 설명하는 beta값임.

B2 : std err로 표기된 0.059 값이 표준 편차
